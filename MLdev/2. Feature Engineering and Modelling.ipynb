```python
import pandas as pd

# Assuming your DataFrame is named df
# Define a function to categorize weather conditions
def categorize_weather(row):
    if row['Rain'] > 0:
        return 'Rainy'
    elif row['Temperature'] < 35:  # Adjust the temperature threshold as needed
        return 'Cold'
    else:
        return 'Sunny'

# Create a new column 'WeatherCondition' based on the categorization
df['WeatherCondition'] = df.apply(categorize_weather, axis=1)
```

```python
# transform and map State
wther_le = LabelEncoder()
wther_labels = wther_le.fit_transform(df['WeatherCondition'])
df['WeatherLabel'] = wther_labels
df
```

```python
# transform and map State
openh_le = LabelEncoder()
openh_labels = openh_le.fit_transform(df['IsOpenHour'])
df['OpenHourLabel'] = openh_labels
df
```

```python
# Convert the 'Timestamp' column to a datetime object
df['Timestamp'] = pd.to_datetime(df['Timestamp'])


df['DayOfWeek'] = df['Timestamp'].dt.dayofweek  # Monday = 0, Sunday = 6
df
```

```python
df_new = df[['Timestamp', 'Binned_Availability', 'WkndLabel', 'PkHrLabel',
            'HolidaysLabel', 'OpenHourLabel', 'Temperature', 'Rain', 'WeatherLabel', 'DayOfWeek']]
df_new
```

```python

# Handling missing values by filling NaN with the mean for numerical columns
numerical_columns = ['Temperature', 'Rain']
categorical_columns = ['WkndLabel', 'PkHrLabel', 'HolidaysLabel', 'OpenHourLabel']

df_new[numerical_columns] = df_new[numerical_columns].fillna(df_new[numerical_columns].mean())

# Handling missing values by filling NaN with the mode for categorical columns
df_new[categorical_columns] = df_new[categorical_columns].fillna(df_new[categorical_columns].mode().iloc[0])

# Checking for and handling infinite values
df_new.replace([np.inf, -np.inf], np.nan, inplace=True)
df_new.dropna(inplace=True)  # Drop rows with any remaining NaN values

# Continue with the rest of your code for model training and evaluation
```

```python
#features = ['WkndLabel', 'PkHrLabel', 'HolidaysLabel', 'Temperature', 'Rain', 'OpenHourLabel']
#features = ['WkndLabel', 'PkHrLabel', 'HolidaysLabel', 'Temperature', 'Rain', 'OpenHourLabel', 'DayOfWeek']
features = ['PkHrLabel', 'HolidaysLabel', 'Temperature', 'Rain', 'OpenHourLabel', 'DayOfWeek']
#features = ['WkndLabel', 'PkHrLabel', 'HolidaysLabel', 'OpenHourLabel', 'Temperature', 'Rain', 'WeatherLabel']
#features = ['WkndLabel', 'PkHrLabel', 'HolidaysLabel', 'OpenHourLabel', 'Temperature', 'Rain', 'WeatherLabel', 'DayOfWeek']

target = 'Binned_Availability'
```

```python
X = df_new[features]
y = df_new[target]
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import numpy as np

from sklearn.svm import SVC, SVR
from xgboost import XGBClassifier, XGBRegressor
from catboost import CatBoostClassifier, CatBoostRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
```

```python
# create model objects for each of the classification model 

models = {
    "RandomForestClassifier": RandomForestClassifier(),
}

accuracy_scores = {}

# perform K-fold cross-validation and run model training and testing
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for model_name, model in models.items():
    accuracies = []  # list to store the accuracy scores
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        # fit the model to the training data
        model.fit(X_train, y_train)
        
        # make predictions on the test set
        y_pred = model.predict(X_test)
        
        # calculate the accuracy score
        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)

    # calculate the average accuracy across all folds
    average_accuracy = np.mean(accuracies)
    accuracy_scores[model_name] = average_accuracy
    
    
```

```python
# print the table of accuracy scores
print("{:<30} {}".format("Model", "Average Accuracy"))
print("-" * 50)
for model_name, accuracy in accuracy_scores.items():
    print("{:<35} {:.4f}".format(model_name, accuracy))
    
```
